### Abstract <a class="anchor" id="abstract"></a>
The structure of ice in the ‘everlasting hunt for new ice phases’1 is the focus of intensive re-search, involving the identification of 20 crystalline and several amorphous phases.2 Here we present a novel view of the structure of the standard hexagonal phase, ice Ih, in the presence of ~1% cubic moiety. A key result is that ice Ih, prepared from pure water, displays net molecular orientation. This gives rise to spontaneous polarization and internal fields of ~4 x 106 Vm-1. This reflects order in the disposition of the protons in ice Ih, a conclusion supported qualitative-ly by earlier computational and experimental work.3,4 Our results require revision of the near century old conventional structure in which water molecular dipoles point in random direc-tions.5 Data are of relevance in atmospheric science, in the solar system, in astrochemistry and in assessing star formation rates,6 the latter relating to the formation of massive galaxies in the early Universe observed recently using the James Webb Space Telescope.7

### Main Text <a class="anchor" id="introduction"></a>

A central issue in the structure of ice Ih is the presence or absence of molecular dipole ordering.8,9 In ice Ih, molecules are positioned on an ordered array of O atoms with molecular dipoles directed along the C2 rotation axes, conventionally pointing in random directions, following the so-called Bernal-Fowler ice rules.10 The spontaneous orientational order reported here, in ice prepared by deposition of pure water vapour at temperatures T ≥127.5 K, represents a weak violation of the ice rules but does not involve sufficient translation of H atoms to form Bjerrum defects. However, the dipole moment of water in the solid is sufficiently large (3.0 D)11,12 to create substantial internal electric fields despite the minor degree of order. Since the order, and hence the extent of polarization and associated electric fields, depend on temperature of ice deposition, the presence of these fields in turn leads to a temperature dependence of the wavelength of excitonic absorption in the vacuum ultraviolet (VUV).13 We demonstrate using VUV spectroscopy that spontaneous polarization leads to an excitonic Stark effect, with spectra showing variation with deposition temperature. Values of spontaneous fields are derived from observed changes in transition energies between 127.5K ≤ T ≤ 140K. Note that a new film is laid down at each temperature and films are not heated or cooled before spectra are taken. In addition, ice films contain ~1% of cubic ice, leading to local stacking disorder (‘sd’). Thus strictly we deal with Isd..14 The phase change between amorphous solid water (ASW), characteristic of deposition temperatures ≤ 125K, and Isd for films prepared at ≥127.5K, may be clearly observed though abrupt shifts in spectral absorption between these temperatures, where the phase change, leading to crystallinity in Isd, is inferred from the presence of excitons for T≥127.5K. In addition we show an electronic spectral signature which may be ascribed to the cubic ice motif in Isd,15,16 enabling the estimation of the 1% contribution of Ic moieties (see below).  
The interest regarding the presence of spontaneous fields in ASW and ice Isd is manifold. Polarization may affect ice particle nucleation rates.17 Polarization in ice Isd can also have an important impact on the rate of formation of young stellar objects, by reducing the degree of ionization in protostellar clouds, hastening the formation of magnetically supercritical cores and thus enhancing the  rate of star formation.6 We note also that local fields in ASW may be sufficiently high to modify chemistry of molecules embedded in ASW on grains in space.18 Electric fields in thin films might be also harnessed to separate charge carriers upon photoionization,19 migrate solvated ions towards an interface or polarise bonds in solvated target molecules to affect the rate of chemical reactions.20 
Spontaneous polarization in water ice in general is a well-known proposition. Ice XI, prepared from dilute KOH, has been suggested as ferroelectric in [21], though brought into doubt in [22]. Violation of the ice rules has also been observed in water ice in hetero-epitaxial ice films. Sum-frequency generation (SFG) spectroscopy revealed spontaneous polarization,23–26 ascribed to the presence of hydronium (H3O+) ions at the interface with the ice film, where ions were formed through hole injection from a Pt(111) substrate. This leads to a strongly non-uniform field. Here by contrast, we infer a uniform field in pure water Isd films with no local polarization arising from the injection of ions from our magnesium fluoride substrate. Kelvin probe measurements report weak fields which permeate ASW condensed on substrates below 125 K.4,10,11 These observations indicate the presence of local fields of >109 Vm-1, created through unbalanced dipoles in pores. The magnitude of these fields may be inferred from theoretical work involving the influence of electric fields on the strength of hydrogen bonds.27,28 
In Fig. 1 we present VUV spectroscopic data for Isd films successively deposited in the temperature range T = 20K to 140K. Spectra are of ~8 nm films or 30 monolayers (ML), obtained at the AU-UV beamline of the ASTRID2 Synchrotron Radiation Facility at Aarhus University: see ‘Methods’. Films are formed by condensation of pure water vapour, from an ambient pressure of ~10-7 mbar, onto a cold magnesium fluoride substrate maintained at a fixed temperature and acting as a VUV window. The VUV flux at the sample is sufficiently low to make spectroscopy a non-intrusive technique, which does not induce chemistry or affect the physical structure of the film. Values of absorbance, n(Io(eV)/I(eV)), vs eV were recorded, where I0(eV) is the transmitted intensity in the absence Isd and I(eV) is the transmitted intensity through the film and substrate. Very high signal-to-noise is achieved through the stability of the ASTRID2 source. This enables the measurement of peak positions to within better than 1 meV in peaks which are >1 eV broad. Three gaussians were fitted to each spectrum, one centred around 8.65 eV, a second centred around 7.98 eV and a third at higher energy to account for an absorption peak with a maximum above the magnesium fluoride cut-off at 10.78 eV (see caption to Fig.1). Earlier reports of VUV spectra of ice may be found in [24,26,29,30]. For comparison, the higher energy peak absorption wavelength for ice at 130K is measured here to be 8.6499 eV and [29] records 8.6503 eV. Excitations in both the higher and lower energy lines belong to the (1b1)(4a1) A 1B1 ← (1b1)2 X 1A1 transition, blue-shifted relative to the equivalent gas phase transition at 7.4 eV.31,32 The absorption at 7.98 eV has not previously been discussed.
Our interpretation is that data in Fig.1 represent the superposition of two independent spectra and we adopt the working hypothesis that these arise from hexagonal Ih and cubic Ic ice. Features at 8.65 eV and 7.98 eV are found for all spectra, where the higher energy feature is associated with hexagonal ice and the lower energy with cubic ice. There is no weakening of the assumed cubic ice feature as we progressively deposit films at 127.5K to 140K. Thus we take the view that over the full temperature range, ice films are composed of both hexagonal and cubic ice moieties in very similar proportion.14,15,33 Material is ASW for T≤125 K and crystalline for T≥127.5K, where the latter is composed of the most common variant of crystalline ice, hexagonal (Ih) ice, with stacking disorder through very restricted cubic ice sequences.14 Assuming the same oscillator strength for both hexagonal and cubic ice, the fits in Fig.1 suggest that the film is of very low cubicity, with ~99% hexagonal ice and 1% cubic. Thus our data refer to spontaneous fields in ice Ih with no contribution to fields from cubic Ic ice since fields are created through the concerted effect of between 5 and 10 ML.13 ASW, formed at T ≤125K, may be seen as nano-crystalline and composed of both hexagonal and cubic ice moieties, but with very little long-range order. For T≥127.5K, ice has large crystalline zones of hexagonal and much smaller zones of cubic moieties.14 This is illustrated in the simulations in Fig. 5 of [14], for growth at 270K, with zones of hexagonal and cubic sequences. At 125K, films become labile as they approach the phase transition, relaxing over a period of 4000 to 5000 seconds. This temporal variation is illustrated in Fig. 2b and c and within fitting errors has the same form for both spectral peaks, displaying time constants of 1622±100 seconds for the higher energy peak and 1539±136 seconds for the lower energy peak. 

With reference to ASW data in Fig. 2a, a qualitative explanation for the variation of absorption wavelengths in Fig.2a rests upon theoretical estimates of the effect of an applied electric field on the energy of hydrogen bonds,18,19 whose results may be summarized as follows. (i) local electric fields must be of the order of several x 109 Vm-1 to have an observable effect on hydrogen bond energy; (ii) such an additional electric field reduces the energy of hydrogen bonds when pointing in the same direction as the dipole moment of the system, as here, up to ~100 K; (iii) a switch in direction of the local electric field, implied by a switch in direction in the average field at >100K, results in a strengthening of the hydrogen bonds, consistent with the non-monotonic behaviour of absorption peak position around 100K seen in Fig. 2a. From hereon we concentrate upon the variation in wavelength of absorption for crystalline ice Isd, that is, for T≥127.5 K.

For Isd with T127.5K, the variation of transition energies with T, shown in Fig. 2a and column 2 in Table 1, for the higher energy feature, may be ascribed to temperature dependent Stark effects.13 The origin of the Stark field lies in the spontaneous orientation of dipoles and relates directly to the spontaneous partial ordering of protons in Ih. Similar shifts in energy to those in Figs. 1 & 2a have been observed in VUV absorption spectra of CO, nitrous oxide, ammonia and methanol and arise through the temperature dependence of the spontaneously created (‘spontelectric’) field.13 We show below that a well-established model for the spontelectric effect, coupled with a simple electrostatic model involving the influence of the spontelectric field on the exciton, reproduces these shifts for T127.5K for both the higher and lower energy features in Fig.1 and Table 1. We use these data to give a quantitative estimate of the degree of orientational order and associated electric fields in highly dominant Ih component of the ice films. Since the spontelectric field requires a sequence of ≥5 to 10 ML to develop,13 the fields here are associated wholly with Ih ice since the cubic moiety represents only ~0.3 ML in total. Thus Ic ice is subjected to the electric field generated within Ih ice, with no spontelectric field generated by Ic ice alone.

The spontelectric field, Esp, acts on Wannier-Mott electron-hole pairs created in the A-state in the lattice of Isd. If the field decreases (say), the field holds the hole and electron less far apart. This in turn causes the energy of the transition to decrease. In this electrostatic model, Coulombic effects yield a Stark energy shift, , for the electronic transition, given by 2 Esp times the physical displacement, r, from the unperturbed separation, r, between the charge and hole, where13,34

r = r[(1-2Espr2)-1/2 – 1]						(1)
In (1), r is the separation of the hole and electron in the hypothetical absence of the field, that is, the diameter of an unperturbed Wannier-Mott exciton and  is the relative permittivity of the medium, reported as 1.716.35 The spontelectric field Esp is a positive quantity since the molecules are on average weakly oriented such that the O-atom points towards the film surface, giving a negative surface potential.3,4  Using (1), the field may be expressed in terms of the energy shift, , as a solution of the equation:
 /{2 r [(1-2 Espr2)-1/2 -1]} - Esp = 0					(2)




from which, by squaring both sides and solving the resulting cubic in E_sp, it follows that:
E_sp= -δ/3r+(δ/3r) [1+6/δϵr]^(1/2) cos⁡[1/3  arccos⁡((1-9/2δϵr)/[1+6/δϵr]^(3/2) )+2πk/3]			(3)                     
with k = 0, 1, 2. Only k = 0 gives a positive root, with k = 1 giving an unphysical negative value of Esp and k = 2 no longer satisfying equation (2), arising from the sign ambiguity introduced by squaring.
Given values of r and , the latter as a function of T, and shown in Fig. 3, Eqn. 3 enables the evaluation of the change of Esp with deposition temperature. Values of  require knowledge of the shift in absorption energy relative to a limiting value E0, to be determined, corresponding to the limit of an infinitely thin film. The evaluation of Esp and degrees of orientation vs T represents an ill-posed problem, whose solution is to appeal to additional knowledge, so-called regularization. This is achieved using a well-established mean-field model of the spontelectric effect, which describes how Esp changes with T.13 Thus we seek values of , via E0, and r, which simultaneously satisfy both Eqn. 3 and the mean-field model for spontelectrics, which we now briefly describe.

As a crystalline layer of water ice accumulates during deposition, incoming molecular dipoles tend spontaneously to orient as they settle into the film. Spontaneous dipole orientation creates a surface potential and an electric field in the film, noting the absence of any free charge. The resulting spa-tially averaged degree of dipole orientation is given by the ratio of the average net component of the dipole moment in the direction z, perpendicular to the plane of the film, relative to the total dipole moment of the species in the solid state, that is <z>/, henceforth m.  The magnitude of m depends on the temperature of deposition, T, which is taken to be the experimentally controlled temperature of the surface. Dipole orientation is a bulk phenomenon and our analysis is therefore restricted to the Ih component of the film, as mentioned above. The mean-field Hamiltonian, E, for the system is characterized by feedback, since m itself depends on the interaction of the molecular dipole with the field, which itself depends on m. We write:

E = Ez = {ES [1+  m2] - EA m}				       (4)
where  is the dipole moment of water in the solid state11,12 and Ez is the mean local field in the z-direction. In eq. 4, the parameter ES represents polarization, dispersion and both covalent and coulombic interactions within the ice, including hydrogen bonding. The term ES  m2 embodies the reluctance of molecules to rotate against one another, which is a determining trait in this strongly hydrogen bonded system. The term EAm gives the value of the spontelectric field in the bulk of the ice. ES and  are free parameters to be determined by fitting to experimental data. The model requires that EA = 4πμ/Ω where Ω = 17.35 Å3 is the molecular volume of water molecules in ice,36 yielding EA = -6.514 x 1010 Vm-1. The standard Langevin relationship37
m = Coth[Ez/T] -  (Ez/T)-1			                              (5)
relates the degree of dipole orientation in the electric field Ez and T. Combining Eqn. 4 and Eqn. 5 gives: 

f(T,m) =  m  =  Coth{[ES[1 +  m2]  - EA m]/T} -  {[ES[ 1 +  m2]  - EA m ]/T}-1         (6)

The task is therefore to find m as a function of T and the corresponding spontelectric field, Esp =         EA m, as a function of T. Values of ES and  are found by fitting so that Eqn. 3 is simultaneously satis¬fied according to a least squares criterion for any chosen exciton size r and limiting value of transition energy, E0. In order to facilitate this we employ the linear inverse form of the Langevin function.38,39 The parameter EA is also allowed to vary since both μ and Ω have associated uncertain-ties, but we find that EA becomes -6.5144 x 1010 Vm-1, remaining very close to the model value. The form of the equations (3) to (6) is such that values of r and E0, and thus , cannot be established independently. However E0 is constrained so that absorption energies are close to those observed spectroscopically.

The energy shifts, , induced in the exciton by the spontelectric field are shown in Fig. 3 as a function of temperature of film deposition, T. These shifts are therefore those which satisfy simultaneously both the electrostatic model described by Eq. 3 and the spontelectric model described by Eq. 6, and the manner in which the electric field varies with T. Extensive numerical experiments show that to fit observations of , E0 lies at 8.475 eV for the higher energy line and 7.818 eV for the lower energy line. Values of r are restricted to 9.75±0.25 nm, a range which gives the best fit between the exciton model and the spontelectric model. The exciton size is therefore found to be much greater than the molecular diameter (0.32 nm), consistent with a Wannier-Mott exciton. Associated values of parameters are EA = -6.5144 x 1010 Vm-1 (see above), ES = 1.8975  x 106 Vm-1,  = 2.8944 x 108. The degree of orientation, m, varies weakly over the temperature range between 127.5K and 140K and falls between -6.12 x 10-5 and -5.87 x 10-5. Note that these values, and the value of EA, are designated negative, due to the direction of orientation of the water molecules. 

Values of m are very small, reflecting a rigid structure, commensurate with a strongly hydrogen bonded system. We show below that the angle of deviation of water molecules from the mean, due to hindered libration, is ~14o from its average degree of orientation of ~cos-1 7 x 10-5 radians (0.004o). Very low values of m nevertheless represent significant bound charge on the surface of the polarized ice, resulting from the high dipole moment of water moieties in ice and the small molecular volume of water. We now turn to a ‘black swan’ which emerges from the above analysis. If we enter the values of EA, ES and  derived above, we find that the total differential of m vs T, derived from eq. 7, becomes infinite at T~140K, subject to the condition that f(T,m) =  m holds simultaneously. This singular behaviour has been described in detail elsewhere,38 where it was shown that a minimum absolute value of m arises where |dm/dT| tends to ∞, given that:

T[2 m ES - EA] (Ez2)-1 -  (/T) (2 m ES - EA) Cosech2 [Ez/T]] -1 = 0		(7)

Using our values of EA, ES and , Eqn. 7 is satisfied for m = -5.88 x 10-5 at 140K, which may be compared with the value of m = -5.87 x 10-5 derived for T = 140K in Fig. 4. Thus the magnitude of degree of orientation is essentially at its minimum value at 140K. Note that a minimum in the varia-tion of m with T may be identified by dm/dT passing through a singularity rather than through set-ting the total differential dm/dT equal to zero. Indeed one may show analytically that dm/dT cannot equal zero for real and finite values of EA, ES and .38 Thus the function governing polarization vs T is continuous but not differentiable at every temperature. This property, by which the total differen-tial of a continuous function cannot be interpreted as its slope, is a long-standing result well-known to mathematicians40 but only first reported in a physical system in [38] and appearing again in the present case of Ih. Our results suggest that the degree of orientation of Ih may rise for T≥140K. This property, that dipoles move to become more oriented close to the desorption temperature, stems from the increased mobility afforded to molecular species in a material close to a phase transition,41 in this case observed as desorption around 140K. Verification might be achievable by using higher chamber pressures to prevent ice sublimation.42 

We now show that the Ih structure allows for highly restricted libration, commensurate with the very small values of dipole orientation, m, derived above. If a molecular dipole is displaced rotationally from some steady state angle, θs, to a new orientation θ, then it will experience a torque tending to return it to θs. This torque arises from the net electric field on the dipole given by the difference be-tween Ez (Eq. 4) for the two orientations represented respectively by θs and the displaced orienta-tion θ. This torque is proportional to the angular acceleration via the molecular moment of inertia, I. Thus the angular acceleration, aθ, is given by (/I)[Ez(θ) sin θ - Ez(θs) sin θs] and the work done in achieving a displacement of θ - θs is the integral over θ of aθ times the moment of inertia. Inserting the expression for Ez and using cos-1 m = θ, yields an expression for the energy expended in rotating a molecule through an angle θ - θs:
[½ EA cos2θ - ES cos θ  - ⅓ES  cos3θ -  θ sin θs (ES – EA cos θs + ES  cos2θs)]		(8)
In the context of the mean-field model, we consider a species with θs given by cos-1 of the values of m shown in Fig. 4. Given azimuthal symmetry, we may then estimate the angle through which a typical molecule may librate at any temperature T by setting Eqn. 8 equal to kT and solving for the resulting two values of θ where k, Boltzmann’s constant, = 1 in atomic units. For example at a depo-sition temperature of T = 130K (say), a molecule oscillates through a total angle of 1.205 degrees, given an average degree of orientation of -6.1 x 10-4 or 0.035 degrees to the horizontal (Fig. 4d). Hence our model implies a very rigid structure allowing for very little librational motion, commensu-rate with a strongly hydrogen bonded system.

### Methods <a class="anchor" id="Methods"></a>
Films of water ice were grown via physical vapour deposition on a cooled MgF2 substrate using the Open University Portable Astrochemistry Chamber (PAC) with a base pressure of 10-9 mbar. The PAC was attached to the AU-UV beamline of the ASTRID2 Synchrotron Radiation Facility at Aarhus University and absorption spectra obtained using linearly polarised tuneable VUV, using the toroidal grating monochromator on this beamline. The high energy cut-off was 10.78 eV (115 nm) via a LiF entrance window and MgF2 exit window, and the resolution was 0.075 nm for all experiments. As discussed in detail in [34], the conditions of our experiment, irradiating an area of 2.5 x 10-5 m2 with 1.2 x 1010 incident photons per second in the bandwidth of the interrogating light, creates an exciton concentration of a few x 107 cm-3. This value results in no contact annihilation and ionization, including effects of diffusion. Photon intensities are such that they contribute negligibly to photodesorption43 or photon-induced crystallization.44
The MgF2 substrate was held in thermal contact with a closed cycle helium cryostat and could be set to any desired temperature 20K and was held constant to ±0.1K, using an Oxford Instruments controller (ITC502). Deionised/ultrapure liquid water was freeze-pump-thaw degassed and its vapour was dosed from the gas phase from a 3 mm diameter capillary positioned 20 mm from the window, yielding close to normal incidence dosing. Films of water were grown between 20 – 140 K, with 140 K marking the highest temperature at which a stable water ice film could form under our pressure conditions. We note that the sticking probability was greatly reduced at the highest temperatures. Film thickness, d, was determined using laser interferometry via reflection of a HeNe laser beam from the surface of the ice during deposition45,46 and is given by d = t l/(4ωn cosθ) where t is the deposition time (/s), λl is the wavelength of the laser = 632.8 nm, ω is the half-period of the oscillation of the optical signal, n is the temperature dependent refractive index of solid H2O46 and θ the angle of incidence of the HeNe laser to the substrate = 20°. For T>125K, this method became increasingly unreliable as the sticking probability became <1. Films were then formed such that the absorbance, log10(I0()/I()) for incident wavelength λ, lay between 0.5 and 0.6. Typically, the rate of deposition was 1.0±0.2  nm s-1.

```python

r_nm_values = np.array([1,2,3,4,5,6,7,8])
l0_values = np.array([144.75])

r_nm_values_2D, l0_values_2D = np.meshgrid(r_nm_values, l0_values)
r_nm_values_3D, l0_values_3D, Efield_3D = np.meshgrid(r_nm_values, l0_values, Temp)

chi2_values = np.zeros(r_nm_values_2D.shape)
figofmerit_values_m = np.zeros(r_nm_values_2D.shape)
figofmerit_values_E = np.zeros(r_nm_values_2D.shape)
all_obs_E_field = np.zeros(r_nm_values_3D.shape)

print('I have a mesh with ', r_nm_values_3D.shape[1], ' values of r; ', r_nm_values_3D.shape[0] ,' values of l0 and ', r_nm_values_3D.shape[2],' spaces for E field values')

eps = (1.70172) #epsilon - usually take (refractive index)**2  - Kofman_2019_ApJ 
s_nm = 0.37 # David provided this value
mu_0_Debye = 3 # solid state dipole moment since this is known for water and avoids us recalculating it with topping equation
Omega_SI = 17.35e-30 # volume in m**3 from saga et al. (normally calculated based on polarizability, but don't do it this way for water ice).

#find the peak fit data
#os.chdir(working_dir)  
df = pd.read_csv(filename+'.txt', delim_whitespace=True, header=(0))
xdataraw = df.iloc[:,0] #Temperature K
ydataraw = df.iloc[:,1] #peak position nm      
        
#set up some graph parameters
T_array=xdataraw.to_numpy()
xmin_mu=min(T_array - 2)
xmax_mu=max(T_array + 2)
xmin=min(T_array - 2)
xmax=max(T_array + 2)
       
#convert units
lambda_nm=ydataraw
lambda_eV=1239.8/lambda_nm

for loop_index_i in range(r_nm_values_2D.shape[0]):
    for loop_index_j in range(r_nm_values_2D.shape[1]):
        #define i and j place in the matrix by scrolling through r and l0 values
        r_nm = r_nm_values_2D[loop_index_i, loop_index_j]
        l0 = l0_values_2D[loop_index_i, loop_index_j]
        r_i=str(r_nm)
        l0_j=str(l0)
        ij_space=('radius: '+r_i+' nm - and l0: '+l0_j+' nm')
        
        l0_eV=1239.8/l0
        DE_eV=(l0_eV-lambda_eV)
        #convert things to atomic units
        T=xdataraw*3.16683e-6
        DE=DE_eV*0.036749405469679
        r=r_nm*18.897259886  #radius in atomic units

        #Calculate the "observed" value of E field, based on an exciton expansion model        
        Esp_au=(1/96)*(
            -((32*DE)/r)
            +(
                (16*(2**(1/3))*DE*(6+DE*eps*r))
                /
                (r**3 *
                ((((DE**2*eps**2 *r**(0.5))*(-9+2*DE*eps*r))+3*(3**(0.5))*(-DE**3 *eps**3 *(32+DE*eps*r*(13+4*DE*eps*r)))**0.5)
                /
                (r**(9/2))
                )**(1/3)
                )
            )
            +
            (
                (8*(2**(2/3))*
                ((((DE**2 *eps**2 *r**(0.5))*(-9+2*DE*eps*r))+3*(3**(0.5))*(-DE**3 *eps**3 *(32+DE*eps*r*(13+4*DE*eps*r)))**0.5)
                /
                 (r**(9/2))
                )**(1/3))
                /
                (eps))
        )

        Esp_Vm=Esp_au*5.14220652e+11
        Esp_Vm_array=Esp_Vm.to_numpy()
        ymin=min(Esp_Vm_array) - 0.01*(Esp_Vm_array[-1])
        ymax=max(Esp_Vm_array) + 0.01*(Esp_Vm_array[-1])
        
        #Convert this E field into degree of dipole alignment
        #first calculate E_a        
        s = (s_nm)*18.897  # estimated layer spacing converted to au
        mu_0 = (mu_0_Debye*0.393456)  #  solid state dipole converted to au

        Omega = Omega_SI*18897259886**3 # convert volume to au
        water_radius_AU = ((3/(4*np.pi))**(1/3))*(Omega**(1/3))
        
        E_A=4*np.pi*mu_0/Omega
        E_A_Vm = E_A*5.14220652e+11      

        degree_of_alignment=Esp_Vm/E_A_Vm
        muz_mu0_array=degree_of_alignment.to_numpy()
        ymin_mu=min(muz_mu0_array) - 0.01*(muz_mu0_array[-1])
        ymax_mu=max(muz_mu0_array) + 0.01*(muz_mu0_array[-1])
       
        #now save E_sp and degree_of_alignment to a file to access later
        length=(len(T_array))
        np.ones(length)
        err_muz_mu0=0.1*(muz_mu0_array)
        err_T=np.ones(length)   #0.1*(T_array)
        
        os.chdir('observed_data')
        data=np.column_stack((weight,muz_mu0_array,err_muz_mu0,T_array,err_T,Esp_Vm_array))#,mu_0,E_A_Vm))
        header='weight,y,sigma_y,Temperature,sigma_T,Observed_field'

        with open('data for_'+name+'.csv', mode='w', ) as newfile: 
            np.savetxt(newfile,data, header=header,delimiter=',',comments='') 
    
        os.chdir('..')  
        
        os.chdir('observed_data')
		
        SpontelData = pandas.read_csv('data for_'+name+'.csv')
        os.chdir('..') 
        #SpontelData.head()
    # Main driver programme for fitting Spontelectric field data
    # 
    # Define several conversion factors to/from atomic units
    # conversion of Temp to a.u. : T [a.u.]= T [K]/factor
        Ttoau=315770.0
    # conversion of D to a.u. : 1 D (=Debye) = factor [a.u.]
        Debye=0.393456
    # Convert el. field in a.u. To V/m : E [a.u.] = E [V/m]/factor
        Etoau=5.142e11
 
        htfc=numpy.array([0.0]*4)
 
    # Find the value of mu for this ice 
        IceMu=mu_0_Debye
    
    # Provide a first guess value for the parameter E_A
        EAnxt=E_A_Vm
    
        htfc[0]=1.0
        htfc[1]=0.0
        htfc[2]=0.0
        htfc[3]=0.0
    
        weight=numpy.array(SpontelData.weight)
        ndata=len(weight)
    # y is the same as <mu_Z>/mu, Temp is the deposition temperature in [K]
    # sig are the 1-sigma measurement errors
    # E observed is the observed Electric field in units [V/m]
        y=numpy.array(SpontelData.y)
        sigy=numpy.array(SpontelData.sigma_y)
        Temp=numpy.array(SpontelData.Temperature)
        sigT=numpy.array(SpontelData.sigma_T)
        Eobs=numpy.array(SpontelData.Observed_field)
    #
    # here comes a block of variable declarations
        whtsum=0.0
        yinvmn=0.0
        hTfunc=numpy.array([0.0]*ndata)
        DerhTfunc=numpy.array([0.0]*ndata)
        DerGT=numpy.array([0.0]*ndata)
        DerGy=numpy.array([0.0]*ndata)
        PartDerv=numpy.array([[0.0]*ndata,[0.0]*ndata,[0.0]*ndata,[0.0]*ndata])
        xdat=numpy.array([[0.0]*ndata,[0.0]*ndata])
        ydat=numpy.array([0.0]*ndata)
        zdat=numpy.array([0.0]*ndata)
        ymodel=numpy.array([0.0]*ndata)
        DParDy=numpy.array([[0.0]*ndata,[0.0]*ndata,[0.0]*ndata])
        DParDT=numpy.array([[0.0]*ndata,[0.0]*ndata,[0.0]*ndata])
        SigPar=numpy.array([0.0]*3)
#
#
# Auxiliary functions f_0, f_1, f_2, and f_3 are needed as input for linear least-squares
# These are named here as PartDerv[0,], ..., PartDerv[3,] because these are also the
# partial derivatives of the function G with respect to T and y.
        for i in range(0,ndata):
            yinvmn+=weight[i]/y[i]
            whtsum+=weight[i]
            hTfunc[i]=htfc[0]+Temp[i]*(htfc[1]+Temp[i]*(htfc[2]+Temp[i]*htfc[3]))
            DerhTfunc[i]=htfc[1]+Temp[i]*(2.0*htfc[2]+3.0*Temp[i]*htfc[3])
            PartDerv[0,i]=(1-y[i])*LangevInv(y[i])
            PartDerv[1,i]=Ttoau*Debye*IceMu*(y[i]-1.0)/Temp[i]
            PartDerv[2,i]=-1.0*PartDerv[1,i]*y[i]
            PartDerv[3,i]=PartDerv[1,i]*y[i]*y[i]*hTfunc[i]
#
# start the iteration to get the parameters, initialise a choice for E_A
        yinvmn=yinvmn/whtsum
        EAcur=1.1*EAnxt
        itcnt=0
        while ((abs(EAnxt/EAcur-1.0)>1.0e-3) and (itcnt<20)) :
            itcnt+=1
            EAcur=EAnxt
            for i in range(0,ndata):
                xdat[0,i]=PartDerv[1,i]
                xdat[1,i]=PartDerv[3,i]
                ydat[i]=-1.0*(PartDerv[0,i]+EAcur*PartDerv[2,i]/Etoau)
#
            Constrout=ConstrLSQ(xdat,ydat,weight)
#
# FPP (02/06/2022): here E_S and zeta always are forced to be >= 0
            EScur=abs(Constrout[0]*Etoau)
            ESzetacur=abs(Constrout[1]*Etoau)
            ratiolog=0.0
            for i in range(0,ndata) :
                Ezmod=(EScur-EAcur*y[i]+ESzetacur*hTfunc[i]*y[i]*y[i])*IceMu*Ttoau*Debye/Etoau/Temp[i]
                ratiolog+=weight[i]*(math.log(abs(EAcur*Langevin(Ezmod)))-math.log(Eobs[i]))
                DerGy[i]=(EScur-ESzetacur*hTfunc[i]*y[i]*(2.0-3.0*y[i])+EAcur*(1.0-2.0*y[i]))*IceMu*Ttoau*Debye/Etoau/Temp[i]
                DerGy[i]+=-1.0*LangevInv(y[i])+(1.0-y[i])*DerLangevInv(y[i])
                DerGT[i]=(EScur*(1-y[i])-EAcur*y[i]*(1-y[i])+ESzetacur*hTfunc[i]*y[i]*y[i]*(1-y[i]))*IceMu*Ttoau*Ttoau*Debye/Etoau/Temp[i]/Temp[i]
                DerGT[i]+=-1.*ESzetacur*DerhTfunc[i]*y[i]*y[i]*(1-y[i])*IceMu*Ttoau*Debye/Etoau/Temp[i]
            ratiolog=ratiolog/whtsum
            if ((Constrout[0]<0) and (EAcur<0) and (yinvmn<0) and (ratiolog>0)) :
                EAnxt=EAcur+2.0*yinvmn*EScur
            else :
                EAnxt=EAcur/math.exp(ratiolog)
            if (itcnt>40) : 
                print ("iteration count exceeded: not converged")
#
# for the calculation of the uncertainty margins for the parameters the partial derivatives below are needed
#
# also calculate the quality of the fit (chi-square) dlsqsum
        dlsqsum=0
        for i in range(0,ndata) :
            ymodel[i]=Langevin((EScur-EAcur*y[i]+ESzetacur*hTfunc[i]*y[i]*y[i])*IceMu*Ttoau*Debye/Etoau/Temp[i])
            DelMod=y[i]-ymodel[i]
            dlsqsum+=weight[i]*DelMod*DelMod/sigy[i]/sigy[i]
        #print(i,weight[i]*DelMod*DelMod/sigy[i]/sigy[i])
            DParDy[0,i]=-1.0*DerGy[i]/PartDerv[1,i]
            DParDy[1,i]=-1.0*DerGy[i]/PartDerv[2,i]
            DParDy[2,i]=-1.0*DerGy[i]/PartDerv[3,i]
            DParDT[0,i]=-1.0*DerGT[i]/PartDerv[1,i]
            DParDT[1,i]=-1.0*DerGT[i]/PartDerv[2,i]
            DParDT[2,i]=-1.0*DerGT[i]/PartDerv[3,i]
#
        dlsqsum=dlsqsum/(whtsum-2)
# calculate the 1-sigma uncertainties for the parameters
        for k in range(0,3):
            for i in range(0,ndata) :
                SigPar[k]+=weight[i]*(DParDy[k,i]*DParDy[k,i]*sigy[i]*sigy[i]+DParDT[k,i]*DParDT[k,i]*sigT[i]*sigT[i]/Ttoau/Ttoau)
            SigPar[k]=Etoau*math.sqrt(SigPar[k]/whtsum)  

        zetacur=ESzetacur/EScur
        ymodel_E=EAcur*ymodel

#Define text to print on each graph
        SigPar[2]=zetacur*math.sqrt(SigPar[2]*SigPar[2]/ESzetacur/ESzetacur+SigPar[0]*SigPar[0]/EScur/EScur)
        string = 'Spont fit parameters'
        string+="\n"+"$E_S =$"+NeaterOut(EScur,SigPar[0])
        string+="\n"+"$E_A =$"+NeaterOut(EAcur,SigPar[1])
        string+="\n"+"$\zeta =$"+NeaterOut(zetacur,SigPar[2])
        string+="\n\n"+"reduced $\chi^2 =$"+'{0:.1f}'.format(dlsqsum)+"\n"

        annotate=(string)

#define residuals for plot (just for visualising)
        res_E=Eobs-ymodel_E
        res_mu=y-ymodel
      
#FigureMu(Temp,ymodel,y,sigT,sigy,xmin_mu,xmax_mu,ymin_mu,ymax_mu,name)
#ymodel=EAcur*ymodel
#FigureE(Temp,ymodel,Eobs,xmin,xmax,ymin,ymax,name)

        fig, ax = plt.subplots(2, 2, sharex=True, gridspec_kw={'height_ratios':[3,1]}, figsize=(10, 5.4))
        fig.subplots_adjust(hspace=0)
        fig.subplots_adjust(wspace=0.02)
        
#plt.title(ij_space)
        ax[0,0].set_xlim(xmin, xmax)
        #ax[0,0].set_ylim(ymin, ymax)
        ax[0,0].scatter(Temp, ymodel_E, marker='+', s=40, color='red', label='model')
        ax[0,0].scatter(Temp, Eobs, marker='o', s=40, color='blue', label='observed')
    #ax.set_yticks(tick)
        ax[0,0].set_xlabel('Temperature / K', fontsize=16, fontname='garamond')
        ax[0,0].xaxis.set_label_position("top")
        ax[0,0].xaxis.tick_top()
        ax[0,0].set_ylabel('Electric Field / Vm$^{-1}$', fontsize=16, fontname='garamond')
        ax[0,0].tick_params(labelsize='large')
        ax[0,0].legend(loc=0, fontsize=14, framealpha=0)
        ax[0,0].text(xmin,ymin,annotate)
    
        ax[0,1].set_xlim(xmin_mu, xmax_mu)
        #ax[0,1].set_ylim(ymin_mu, ymax_mu)
        ax[0,1].errorbar(Temp, ymodel, markersize=10, xerr=sigT, yerr=sigy, fmt='.', color='red', label='model')
        ax[0,1].scatter(Temp, y, marker='^', s=50, color='blue', label='observed')
        ax[0,1].set_xlabel('Temperature / K', fontsize=16, fontname='garamond')
        ax[0,1].xaxis.set_label_position("top")
        ax[0,1].xaxis.tick_top()
        ax[0,1].yaxis.set_label_position("right")
        ax[0,1].yaxis.tick_right()
        ax[0,1].set_ylabel('Dipole orientation / $\mu_Z/\mu$', fontsize=16, fontname='garamond')
        ax[0,1].tick_params(labelsize='large')
        ax[0,1].legend(loc=0, fontsize=14, framealpha=0)
        ax[0,1].text(xmin_mu,ymin_mu,ij_space+'\n Calculated value of E_A is '+str(round(E_A_Vm/1E9,3))+' x 10^9 V/m')
#ax[0,1].text(20,0,legend_mu)
    
        ax[1,0].scatter(Temp, res_E, marker='+', s=40, color='red', label='observed')
        ax[1,0].axhline(y = 0.0, color = 'g', linestyle = '-')
        ax[1,0].set_xlabel('Temperature / K', fontsize=16, fontname='garamond')
        ax[1,0].set_ylabel('res', fontsize=16, fontname='garamond')
    
        ax[1,1].scatter(Temp, res_mu, marker='+', s=40, color='red', label='observed')
        ax[1,1].yaxis.set_label_position("right")
        ax[1,1].axhline(y = 0.0, color = 'g', linestyle = '-')
        ax[1,1].yaxis.tick_right()
        ax[1,1].set_xlabel('Temperature / K', fontsize=16, fontname='garamond')
        ax[1,1].set_ylabel('res', fontsize=16, fontname='garamond')
       

        
        short_ij=('r_'+str(r_nm)+'_lambda0_'+str(l0))
        os.chdir('Spont fit')
        plt.savefig('Spont fit for'+short_ij+'.png')    
    
        zetacur=ESzetacur/EScur
        
        SigPar[2]=zetacur*math.sqrt(SigPar[2]*SigPar[2]/ESzetacur/ESzetacur+SigPar[0]*SigPar[0]/EScur/EScur)
        string = name
        string+="<br>"+"$E_S =$"+NeaterOut(EScur,SigPar[0])
        string+="<br>"+"$E_A =$"+NeaterOut(EAcur,SigPar[1])
        string+="<br>"+"$\zeta =$"+NeaterOut(zetacur,SigPar[2])
        string+="<br><br>"+"reduced $\chi^2 =$"+'{0:.1f}'.format(dlsqsum)+"<br>"
        #display(Markdown(string)) #Use this line to see the individual results for fitting parameters per dataset
    
        ES=NeaterOut(EScur,SigPar[0])
        EA=NeaterOut(EAcur,SigPar[1])
        zeta=NeaterOut(zetacur,SigPar[2])
        os.chdir('..')  
        
        chi2_values[loop_index_i, loop_index_j] = dlsqsum
        ES_test=(EScur/5.14220652e+11)
        EA_test=(EAcur/5.14220652e+11)
        zeta_test=(zetacur)

        m_check= - (T/(mu_0*((-EA_test*muz_mu0_array)+ES_test*(1+(muz_mu0_array**2)*zeta_test)))) + 1/(np.tanh((mu_0*(-EA_test*muz_mu0_array+ES_test*(1+(muz_mu0_array**2)*zeta_test)))/T))
        E_check = np.sum(((ymodel_E-Eobs)**2)**0.5)
        figofmerit_E = E_check #some estimate of errors based on least squares regression
        figofmerit_m = np.sum(((m_check-muz_mu0_array)**2)**0.5) 
        
        figofmerit_values_m[loop_index_i, loop_index_j] = figofmerit_m
        figofmerit_values_E[loop_index_i, loop_index_j] = figofmerit_E
                
        #Also, save the plots of each obs E field, because this changes a lot as r and l0 vary
        
        for loop_index_z in range(r_nm_values_3D.shape[2]):    
            Efield = Eobs[loop_index_z]   
            all_obs_E_field[loop_index_i, loop_index_j, loop_index_z] = Efield                          
        
```
