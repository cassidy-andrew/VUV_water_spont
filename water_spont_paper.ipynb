### Abstract <a class="anchor" id="abstract"></a>
We describe a novel aspect of the structure of the hexagonal phase of crystalline water ice, using vacuum-ultraviolet absorption spectroscopy as a diagnostic tool. The key result is that hexagonal ice ($I_h$), prepared from condensation of pure water vapour, displays net molecular orientation. This reflects order in the disposition of the protons in ice $I_h$ and our results require revision of the near century old conventional structure in which water molecular dipoles point in random directions in hexagonal ice. Orientation gives rise to spontaneous polarization and internal fields of $~4 \times 10^6\,$ V m$^{-1}$ in films of ice $I_h$ tens of nanometres thick, condensed on an insulating MgF$_2$ substrate. Data are in the form of Stark shifts in absorption spectra, with no free charges or external electric fields introduced in the experiment.

### Introduction <a class="anchor" id="introduction"></a>

Electric fields can change the chemical structure, electronic behaviour and motion of molecules.[1](#ref_1) The effect of these fields has become a new frontier in areas ranging from catalysis, across biology, and to the engineering of electronic devices. For example, in biology, effects are often based on the natural ability of enzymes to generate local electric fields around molecular structures to catalyse specific processes. In the present work, we report on the inherent ability of nanoscale films of crystalline hexagonal water ice to spontaneously achieve a structure by which dipoles orient and produce an electric field internal to the solid. 

Our results challenge the concept that solid crystalline water is always a proton disordered material. Hexagonal water ice, $I_h$, represents *“ordinary”* crystalline ice in our everyday experience. A central issue in the structure of ice $I_h$ is the presence or absence of molecular dipole ordering.[2](#ref_2),[3](#ref_3),[4](£ref_¤) In ice $I_h$, molecules are positioned on an hexagonal array of $O$ atoms with molecular dipoles directed along the $C2$ rotation axes, conventionally pointing in random directions giving no dipole order, following the Bernal-Fowler ice rules.[5](#ref_5) The spontaneous orientational order reported here, in crystalline ice, therefore represents a weak violation of the ice rules but  does not involve sufficient translation of H atoms to form Bjerrum defects. The dipole moment of water in the solid is sufficiently large, 3.0$\,$D[6](#ref_&),[7](#ref_7) to create substantial internal electric fields despite the minor degree of order.

Spontaneous polarization in non-crystalline water ice, formed below 127.5$\,$K, is a known proposition. Onsager et al. observed that condensing samples of pure water ice at 78$\,$K, to form amorphous solid water (ASW), produced electrical effects.[8](#ref_8) Subsequent Kelvin probe measurements concurred, and reported weak fields permeating ASW, formed by condensation on substrates at $\leqq$125$\,$K.[6](#ref_6),[7](#ref_7),[9](#ref_9) These observations indicate a contribution from the presence of local fields of $>$10$^9$$\,$V m$^{-1}$, created through unbalanced dipoles in pores [6](#ref_6),[10](#ref_10) and sufficient to influence the hydrogen bond energy.[11](#ref_11),[12](#ref_12) Such fields can be recreated using molecular dynamics simulations.[13](#ref_13)

Previous observations of polarized crystalline water ice, as here, have been associated with injection of ions into the ice at the ice-substrate boundary. For example, Ice *XI*, prepared from dilute KOH, has been suggested as ferroelectric in [14](#ref_14), though brought into doubt in [15](#ref_15). Violation of the ice rules has also been observed in water ice in hetero-epitaxial ice films. Sum-frequency generation (SFG) spectroscopy revealed spontaneous polarization,[16](#ref_16),[17](#ref_17) stabilized by the presence of hydronium (H$_3$O$^+$) ions at the interface with the ice film, where ions were formed through hole injection from a Pt(111) substrate. This leads to a strongly non-uniform field. Here by contrast, we use a substrate with a high band gap, MgF$_2$, from which we may assume at our low temperatures that there is no injection of ions and thus we maintain pure water ice.

In the present work we report vacuum ultra-violet (VUV) absorption spectra for water ices condensed across a temperature range from 20$\,$K to 140$\,$K. This temperature range crosses the amorphous to crystalline phase transition for water ice at ~127.5$\,$K. Below the phase transition, the behaviour of VUV spectra concurs with what is known about the collapse of pores in films of ASW. Above the crystallisation temperature, films demonstrate an excitonic Stark effect,[18](#ref_18) with absorption wavelengths showing correlation with deposition temperature. We have previously demonstrated, for films of carbon monoxide, nitrous oxide and ammonia, that this Stark effect in VUV spectroscopy can be used to quantify the rate of change in the degree of polarisation with condensation temperature. Moreover this rate of change can be understood by a model for the spontelectric effect.[18](#ref_18) This model, introduced in more detail below, ascribes polarization to a metastable equilibrium established between water dipoles, which align with the intrinsic electric field that arises in condensed molecular films, and thermally induced disorder. In short, the spontelectric model describes spontaneous polarization as a homogeneous bulk effect. Hence, we report a metastable polarised phase of crystalline water ice.



### Methods <a class="anchor" id="methods"></a>

Films of pure water ice were condensed, from a partial pressure of $~$10$^{-7}$$\,$mbar, onto a cold MgF$_2$ substrate maintained at a chosen fixed temperature and acting as a VUV window, held inside a chamber with a base pressure <2 x 10$^{-8}$$\,$mbar. Deionised water was purified using multiple freeze-pump-thaw cycles prior to introduction to the vacuum measurement chamber, using a high precision variable leak valve attached to a tube of 6$\,$mm diameter that lay 25$\,$mm from the substrate.  Film thicknesses were quantified using a laser interferometer technique.[19](#ref_19) Note that a new film was laid down at each temperature and films were not heated or cooled before spectra were taken. Temperatures were measured using a Si-diode located close to the MgF$_2$ substrate and the temperature was controlled by heating the cryo finger via a resistive heater. Data were taken on the UV beamline on the ASTRID2 facility at Aarhus University, where the high stability of the VUV source allows for the very accurate spectral measurements necessary for this experiment. The low VUV flux did not induce chemistry or affect the physical structure of the film.[20](#ref_20) Values of absorbance, $ln(I_0(eV)/I(eV))$, *vs* eV were recorded, where $I_0(eV)$ is the transmitted intensity in the absence of a water film and $I(eV)$ is the transmitted intensity through the film and substrate. 

### Results <a class="anchor" id="results"></a>

The VUV spectrum for a water ice film grown at 130$\,$K is shown in [Fig. 1](#figure_1)a. Earlier reports of VUV spectra of ice may be found in [21](#ref_21),[22](#ref_22),[23](#ref_23),[24](#ref_24). Representative spectra recorded of water ice films deposited at 125$\,$K – 140$\,$K, in the spectral region of interest, are shown in [Fig. 1](#figure_1)b. Three gaussians were fitted to each spectrum in [Fig. 1](#figure_1)a: a background peak to account for an absorption with a maximum above the MgF$_2$ cut-off at 10.78$\,$eV; a higher energy absorption peak around 8.65$\,$eV and a lower energy absorption peak around 7.98$\,$ eV. Our working hypothesis is that these two absorptions arise from the (1b$_1$)(4$_{a1}$) A $^1$B$_1$ $\leftarrow$ (1b$_1$)$^2$ X $^1$A$_1$ transition, blue-shifted relative to the equivalent gas phase transition at 7.4$\,$eV [25](#ref_25),[26](#ref_26), in hexagonal $I_h$ ice and cubic $I_c$ ice respectively. Thus we take the view that over the full temperature range, ice films are composed of both hexagonal and cubic ice moieties giving stacking disordered ice, $I_{sd}$.[27](#ref_27),[28](#ref_28),[29](#ref_29),[30](#ref_30) Assuming the same oscillator strength for both hexagonal and cubic ice, films are ~99% hexagonal and 1% cubic. Below, we demonstrate that the shift in peak positions with temperature arises from the Stark effect, with the macroscopic electric field intrinsic to the long-range ordered arrangement of dipoles within the condensed water ice film. Thus, our data refer to electric fields that are intrinsic to  hexagonal $I_h$ ice, with no contribution to these fields from cubic $I_c$ ice, since spontaneous fields are created through the concerted effect of at least 5 to 10 ML of material.[18](#ref_18) Thus $I_c$ ice is subjected to the electric field generated within $I_h$ ice, with no spontelectric field generated by $I_c$ ice alone. Positions for the peak centred at 8.65$\,$eV are shown as a function of sample deposition temperature in [Fig. 2](#figure_2). Both peak positions for crystalline films are listed in [Table 1](#table_1). 

Material is ASW for $T\, \leq \, $125$\,$ K and crystalline for $T\, \geq\,$ 127.5$\,$ K. ### Data analysis <a class="anchor" id="data_analysis"></a>

For $I_{sd}$ with $T\, \geq \,$127.5$\,$K, the variation of transition energies with temperature, shown in [Fig. 2](#figure_2) and columns 2&3 in [Table 1](#table_1), may be ascribed to temperature dependent Stark effects.[18](#ref_18) The origin of the electric field responsible for the shift in the energy of the excitations, lies in the spontaneous orientation of dipoles and relates directly to the spontaneous partial ordering of protons in $I_h$ ice. Similar shifts in energy to those in [Figs. 1](#figure_1) & [2](#figure_2) have been observed in VUV absorption spectra of CO, nitrous oxide, ammonia and methanol and arise through the temperature dependence of the spontaneously created electric field - the *"spontelectric field"*.[18](#ref_18) We show below that a well-established model for the spontelectric effect, coupled with a simple electrostatic model involving the influence of the spontelectric field on the exciton, reproduces these shifts for T$\geq$127.5$\,$K for both the higher and lower energy features in [Fig. 1](#figure_1) and [Table 1](table_1). We use these data to give a quantitative estimate of the degree of orientational order and associated electric fields in the dominant $I_h$ component of the ice films. 

The spontelectric field, $E_{sp}$, acts on excitons, specifically Wannier-Mott electron-hole pairs, created in the A-state in the lattice of $I_{sd}$. If the ice is formed at higher temperature, say, the strength of the spontelectric field, $E_{sp}$, is lower and holds the hole and electron less far apart. This in turn causes the energy of the transition to decrease and is observed as the absorption peak red shifting. In the hypothetical absence of any intrinsic electric field in the ice film, the speration between the charge and hole within the exciton after excitation is represented by $r$. Dipoles orient at all temperatures for crystalline ice, and the resulting intrinsic electric fields produce  Coulombic effects that yield a Stark energy shift in the observed energy of the excitation. We describe this energy shift as $\delta$, and it is given by  $2 E_{sp}$ times the physical displacement, $\Delta r$, from the unperturbed separation, $r$, where [18](#ref_18),[20](#ref_20)

[Eqn. 1]<a class="anchor" id="equation_1"></a>
\begin{align}
\mathrm{(1)} \quad  \Delta r = r [ ( 1 - 2 E_{sp} \epsilon r^2 ) ^ {-1/2}  - 1]	                        
\end{align}

with $\epsilon$ as the relative permittivity of the medium, reported as 1.716.[31](#ref_31) The spontelectric field, $E_{sp}$, is a positive quantity since the molecules are on average weakly oriented such that the O-atom points towards the film surface, giving a negative surface potential.[9](#ref_9),[13](#ref_13)  Using [Equation 1](#equation_2), the strength of the intrinsic electric field may be expressed in terms of the energy shift observed in spectra, $\delta$, as a solution of the equation:

[Eqn. 2]<a class="anchor" id="equation_2"></a>
\begin{align}
\mathrm{(2)} \quad \frac{\delta }  { 2r[(1-2E_{sp}\epsilon r^2 ) ^{-1/2} -1]  } -  E_{sp} = 0
\end{align}

from which, by squaring both sides and solving the resulting cubic in $E_{sp}$, it follows that:

[Eqn. 3]<a class="anchor" id="equation_3"></a>
\begin{align}
\mathrm{(3)} \quad E_{sp} = - \frac{\delta}{3r} +  \left( \frac {\delta }{3 r} \right) \left[ 1 + \frac{6}{ \delta \epsilon r} \right]^{\frac{1}{2}} \cos \left[ \frac{1}{3} \arccos \left(  \frac {1 - \frac{9}{2\delta \epsilon r} } {\left[1 + \frac {6}{\delta \epsilon r}  \right] ^ { \frac {3}{2}}} \right) + \frac {2 \pi k}{3} \right]
\end{align}

with k$\,$=$\,$0, 1, 2. Only k$\,$=$\,$0 gives a positive root, with k$\,$=$\,$1 giving an unphysical negative value of $E_{sp}$ and k$\,$=$\,$2 no longer satisfying equation (2). This arises from the sign ambiguity introduced by squaring.

Given values of r and $\delta$, the latter as a function of temperature, and shown in [Fig. 3](#figure_3), [Eqn. 3](#equation_3) enables the evaluation of the change of $E_{sp}$ with deposition temperature. Values of $\delta$ require knowledge of the shift in absorption energy relative to a hypothetical limiting value in absence of any intrinsoc electric field in the ice film, $E_0$. Here, we constrain $E_0$, such that the absorption peak position occurs at the limit of an infinitely thin film.

The evaluation of $E_{sp}$ and degrees of orientation *vs* T represents an ill-posed problem, whose solution is to appeal to additional knowledge, so-called regularization. This is achieved using a well-established mean-field model of the spontelectric effect, which describes how $E_{sp}$ changes with temperature.[18](#ref_18) Thus we seek values of $\delta$, via $E_0$ and r, which simultaneously satisfy both [Eqn. 3](#equation_3) and the mean-field model for spontelectrics, which we now briefly describe.


As a crystalline layer of water ice accumulates during deposition, incoming molecular dipoles tend spontaneously to orient as they settle into the film. Spontaneous dipole orientation creates a surface potential and an electric field in the film, noting the absence of any free charge. The resulting spatially averaged degree of dipole orientation is given by the ratio of the average net component of the dipole moment in the direction $z$, perpendicular to the plane of the film, relative to the total dipole moment of the species in the solid state, that is $<\mu_z>/\mu$, henceforth $m$.  The magnitude of $m$ depends on the temperature of deposition, $T$, *i.e.*, the temperature of the surface. Dipole orientation is a bulk phenomenon and our analysis is therefore restricted to the $I_h$ component of the film, as mentioned above. The mean-field Hamiltonian, $\mathfrak{E}$, for the system is characterized by feedback, since $m$ itself depends on the interaction of the molecular dipole with the field, which itself depends on $m$. We write:

[Eqn. 4]<a class="anchor" id="equation_4"></a>
\begin{align}
\mathrm{(4)} \quad \mathfrak{E} = \mu E_z = \mu \{E_s[ 1 +  \zeta m ^2] - E_A m\}
\end{align}

where $\mu$ is the dipole moment of water in the solid state[6](#ref_6),[7](#ref_7) and $E_z$ is the mean local field in the $z$-direction. In [Eqn. 4](#equation_4), the parameter $E_S$ represents polarization, dispersion and both covalent and coulombic interactions within the ice, including hydrogen bonding. The term $\mu E_S \zeta m^2$ embodies the reluctance of molecules to rotate against one another, which is a determining trait in this strongly hydrogen bonded system. The term $E_Am$ gives the value of the spontelectric field in the bulk of the ice. $E_S$ and $\zeta$ are free parameters to be determined by fitting to experimental data. Analysis [32](#ref_32) shows that the model requires that $E_A = 4\pi \mu / \Omega$ where $\Omega$ = 17.35$\,$Å$^3$ is the molecular volume of water molecules in ice,[33](#ref_33) yielding $E_A$ = -6.514$\,\times \,$10$^{10}$$\,$Vm$^{-1}$. The temperature of the substrate induces disored into the dipole orientation, and this is modeled using the Langevin relationship. Combining [Eqn. 4](#equation_4) with the standard Langevin relationship [34](#ref_34) yields: 

[Eqn. 5]<a class="anchor" id="equation_5"></a>
\begin{align}
\mathrm{(5)} \quad f(T,m) = m = \coth \left\{ \frac{ \mu [E_S [1 + \zeta m^2 ] - E_A m] } { T } \right\} - \left\{ \frac {\mu[E_S [1 + \zeta m^2 ] - E_A m } { T } \right\} ^{-1}
\end{align}

The task is therefore to find $m$ and the corresponding spontelectric field, $E_sp$ = $E_A m$, as a function of temperature. Thus, the two means of solving for $E_{sp}$, namely *via* the spontelectric model contained in [Eqn. 5](#equation_5) and *via* the exciton moedl contained by [Eqn. 3](#equation_3), provide for a regularized least squares approach. Values of $E_S$ and $\zeta$ are found by fitting [Eqn. 5](#equation_5), so that [Eqn. 3](#equation_3) is simultaneously satisfied according to a least squares criterion for any chosen exciton size $r$ and limiting value of transition energy, $E_0$. In order to facilitate this we employ the linear inverse form of the Langevin function.[35](#ref_35),[36](#ref_36) The parameter $E_A$ is also allowed to vary since both $\mu$ and $\Omega$ have associated uncertainties, but we find that $E_A$ becomes -6.5144$\,\times \,$10$^{10}$$\,$Vm$^{-1}$, remaining very close to the model value. The form of the [Eqns 3](#equation_3) to [5](#equation_5) is such that values of $r$ and $E_0$, and thus $\delta$, cannot be established independently. However, $E_0$ is constrained, first by observations, such that $E_0$ is lower in energy than $E$ and such that $\delta$ ([Fig. 3](#figure_3)) is equal for both the high energy and low energy transitions, and also by regularization using the spontelectric model. Only values of $\delta$ that allow the spontelectric model to recreate a monotonic change with temperature are permissible.

These constraints and numerical experiments show that $E_0$ lies at 8.475$\,$eV for the higher energy line and 7.818$\,$eV for the lower energy line. The corresponding values of $\delta$ are shown in [Fig. 3](#figure_3) and the resulting fits are shown in [Fig. 4](#figure_4), with values from the exciton model and the spontelectric model listed in [TABLE 1](#table_1). Values of $r$, which is the most influential parameter in this context, are restricted to 9.75$\pm$0.25$\,$nm, a range which gives the best fit between the exciton model and the spontelectric model, see [Fig. 4b](#figure_4). The exciton size is therefore found to be much greater than the molecular diameter (0.32$\,$nm), consistent with a Wannier-Mott exciton. Associated values of parameters in the spontelectric model are $E_A$ = -6.5144$\,\times \,$10$^{10}$$\,$Vm$^{-1}$ (see above), $E_S$ = 1.8975$\,\times \,$10$^6$$\,$Vm$^{-1}$, $\zeta$ = 2.8944 x 10$^8$. The degree of orientation, $m$, varies weakly over the temperature range between 127.5$\,$K and 140$\,$K and falls between -6.12$\,\times \,$10$^{-5}$ and -5.87$\,\times \,$10$^{-5}$. Note that these values, and the value of $E_A$, are negative, due to the direction of orientation of the water molecules. 

Values of $m$ are very small, reflecting a rigid structure, commensurate with a strongly hydrogen bonded system. Very low values of $m$ nevertheless represent significant bound charge on the surface of the polarized ice, resulting from the high dipole moment of water moieties in ice and the small molecular volume of water. These together yield the large quoted value of $E_A$ and give fields, $E_{sp}$ =  $mE_A$, of the order of 6$\,\times \,$10$^{-5}$$\,\times \,$6.514$\,\times \,$ 10$^{10}$$\,$Vm$^{-1}$ corresponding to 1.30$\,$mV per ML. Thus the surface of the 30$\,$ML films studied here take on a negative potential of typically 39$\,$mV or 0.039$\,$Cm$^{-2}$.

### Discussion <a class="anchor" id="Discussion"></a>

Onsager *et al.*,[8](#ref_8) in agreement with Kutzner,[37](#ref_37) ascribe the bulk polarization of water ice films wholly to dipole orientation, templated by the orientation that molecules achieve immediately after landing on the substrate. Subsequent work [6](#ref_6),[7](#ref_7),[9](#ref_9),[10](#ref_10) has shown that there is a significant contribution from pores in ice formed by deposition at 77$\,$K. Be that as it may, the data of Onsager *et al.* [8](#ref_8) based upon measurements of a thermally stimulated current measured at a copper substrate, give an electric field of 3.95$\,\times\,$10$^6\,$V$\,$m$^{-1}$ for ice deposited at liquid nitrogen temperature (77$\,$K). At this temperature there exists (i) the influence on the wavelength of absorption of very high local fields of the order of 10$^9\,$V$\,$m$^{-1}$ arising from pores in the ice and which affect the hydrogen bond energy[11](#ref_11),[12](#ref_12) and thus spectral transition energies and (ii) the presence of non-local spontelectric fields (termed in [8](#ref_8) as a ‘natural electret’) affecting the energy of excitons, themselves arising from the nanocrystalline nature of ASW. Note that similar fields to the latter have been observed in low temperature ammonia ices which are nanocrystalline.[38](#ref_38) The spontelectric fields which we derive for $I_{sd}$ ice at higher deposition temperatures relevant to crystalline ice, appear similar in value to the field derived at this one temperature of 77$\,$K from the data of Onsager *et al.* collected by quite different means and relevant to ASW.

We note that the nature of the surface upon which spontelectric films are deposited has little or no bearing on the values of film polarization.[39](#ref_39) Thus, water ice deposited on substrates other than MgF$_2$, for example, interstellar grains composed of silicates, would yield the same spontelectric fields as reported here. Turning specifically to the interstellar medium, polarization in ice can have an important impact on the rate of formation of young stellar objects, by reducing the degree of ionization in protostellar clouds, hastening the formation of magnetically supercritical cores and thus enhancing the rate of star formation.[40](#ref_40) This may be relevant to the formation of massive galaxies in the early Universe observed recently using the James Webb Space Telescope.[41](#ref_41) It is known that water molecules in the ISM form in situ on the surface of cold dust grains, with temperatures of 10$\,$K, via reactions between mobile H atoms and O atoms, and that this eventually leads to the build-up of water ice. On a timescale of $\sim\,$105$\,$years, these reactions can form a typical coverage of 100 monolayers.[42](#ref_42) This first generation of ASW ice on grains in the ISM is created under very different conditions from those in our current experiment and is unlikely to be spontelectric. However, interstellar water ice is evaporated into the gas phase every 2$\,\pm\,$0.5$\,$Myr via interactions between dust grains and passing shocks.[40](#ref_40) This results in water concentrations $\sim\,$10$^{-4}$ of the molecular hydrogen number density, several orders of magnitude greater than any yield from gas phase water production. These stochastic high gas phase abundances are, for example, required to account for both the brightness and range of frequencies of water masers found in star-forming regions and around late type stars.[43](#ref_43) Thus a second, and possible subsequent, generation of water ice may indeed be spontelectric, formed as water ice layers recondense on cold dust grains following the passage of a shock.

In addition, $I_{sd}$ ice is posited to be found in the Earth’s mesosphere and stratosphere, in Mars, Venus and elsewhere in the solar system.[29](#ref_29) Should such $I_{sd}$ ices be spontelectric, the consequent bound charge at the surface of ice particles in the solar systems may inhibit particle coagulation and therefore rocky planet formation, and ice particle nucleation rates in general.[44](#ref_44),[45](#ref_45)


Although some crystalline water ice (*i.e.* at T$\,\geq\,$125$\,$K) has been described as porous,[46](#ref_46),[47](#ref_47),[48](#ref_48),[49](#ref_49) the structural properties of water ice are highly sensitive to its formation conditions and thermal history. Even within a single formation method such as vapour deposition, parameters including deposition temperature,[50](#ref_50) dosing method,[51](#ref_51),[52](#ref_52),[53](#ref_53) thermal history,[54](#ref_54) and film thickness [50](#ref_50),[54](#ref_54) can significantly affect porosity. These dependencies show that porosity cannot be generalised across vapour deposited ices without accounting for growth conditions and thermal history. Our deposition conditions and film thickness most closely match those of Kimmel *et al.*,[54](#ref_54) suggesting that our water ices grown above the transition temperature of 125$\,$K are more likely non-porous crystalline ices, see supporting information. In contrast, porous crystalline ices have typically been produced via fundamentally different formation routes, such as freezing of liquid water,[47](#ref_47) or for much greater film thicknesses  and different thermal histories.[48](#ref_48),[50](#ref_50),[55](#ref_55).


### Conclusion<a class="anchor" id="Conclusion"></a>

With this report, we have described polarization in crystalline ice $I_h$ as a bulk property that arises following the collective alignment of dipoles throughout the water film. Future work will aim at synthesising and characterising polarization in different ice polymorphs by controlling the temperature and pressure during film growth.  The interest regarding the presence of spontaneous fields in crystalline ice is manifold. Polarization will affect ice particle nucleation rates,[44](#ref_44) and can impact on the rate of formation of young stellar objects in the ISM. Electric fields in thin films might be also harnessed to separate charge carriers upon photoionization,[56](#ref_56), migrate solvated ions towards an interface or polarise bonds in solvated target molecules to affect the rate of chemical reactions.[57](#ref_57)

### Abstract <a class="anchor" id="abstract_sup"></a>

We describe a novel aspect of the structure of the hexagonal phase of crystalline water ice, using vacuum-ultraviolet absorption spectroscopy as a diagnostic tool. The key result is that hexagonal ice ($I_h$), prepared from condensation of pure water vapour, displays net molecular orientation. This reflects order in the disposition of the protons in ice $I_h$ and our results require revision of the near century old conventional structure in which water molecular dipoles point in random directions in hexagonal ice. Orientation gives rise to spontaneous polarization and internal fields of $~4 \times 10^6\,$ Vm$^{-1}$ in films of ice $I_h$ tens of nanometres thick, condensed on an insulating MgF$_2$ substrate. Data are in the form of Stark shifts in absorption spectra, with no free charges or external electric fields introduced in the experiment.

### Sup Mat.2 - Details on peak fitting routine <a class="anchor" id="sup_2"></a>

At a given film growth temperature, several spectra were collected following incremental doses to increase water ice film thickness. Typically, this led to four or five spectra at each growth temperature. All spectra were fit simultaneously using a linear combination of four Gaussians. Three of these Gaussians represent features resulting from the water film and are shown in [Sup Mat Fig. 1](#sup_1) above. The fourth Gaussian was included to account for a small peak at 6.5$\,$eV. This fourth peak, labelled “fourth peak” in the fitting routine, was observed with identical parameters in each spectrum, *i.e.*, the peak did not grow with increasing water doses and did not shift in position as a function of film thickness nor changes to film growth temperature. This peak had an absorbance below 0.05 in all spectra. 

The equation used to describe a Gaussian peak amplitude, *i.e.*, absorbance (Abs), at a given photon energy, $x$ (eV), was

[Sup Mat - Eqn. 1]<a class="anchor" id="sup_equation_1"></a>
\begin{align}
\mathrm{(Sup Mat 1)} \quad Abs = Ae^{\frac {x-x0^2}{2sig^2}}
\end{align}

where $A$ is the peak amplitude, $x0$ is the peak position and $sig$ is the peak width (*i.e.*, FWHM). This gives four peaks each with 3 fitting parameters, giving 12 variables for each spectrum. These variables were constrained with bounding values to allow the fitting routine to converge. These bounds are described in [Sup Mat Table 1](#sup_table_1). The Gaussian labelled *“BkGd peak”* in [Sup Mat Table 1](#sup_table_1) represents the rising continuum towards the first ionisation energy for water ice and at least in gas phase spectra continues to grow in absorbance at energies that are outside the range of our experiment. The Gaussians labelled *“BkGd peak”* and *“fourth peak”* were tightly constrained in position but the peaks that are the focus of this investigation, *“Higher Energy Peak”* and *“Lower Energy Peak”* were fit with parameters that lie well within the bounds, *i.e.*, the position and width of these peaks were not constrained by the choice of bounds. This is more easily visualised by exploring the executable paper described in [section 4](#sup_4).

Data indicate a phase change between 125$\,$K and 127.5$\,$K. Peak positions for the high energy (black dots - b) and low energy (red triangles - c) showing how peaks evolve with time at 125$\,$K. Exponential fits to the time evolution are shown. Errors, marked as vertical lines in (a), are 1$\,$SD on the Gaussian fit and amount to systematic errors of up to 3.$\,$meV and $\,$meV for the higher and lower energy peaks respectively._

Material is ASW for T$\leq$125$\,$K and crystalline for T$\geq$127.5$\,$K. At 125$\,$K, films become labile as they approach the phase transition, relaxing over a period of thousands of seconds. This temporal variation is illustrated in [Sup Mat Fig. 2b and c](#sup_figure_2) and within fitting errors has the same form for both spectral peaks, displaying time constants of 1622$\pm$100$\,$seconds for the higher energy peak and 1539$\pm$136$\,$seconds for the lower energy peak. 

Kelvin probe measurements, [1](#sup_ref_1),[2](#sup_ref_2),[3](#sup_ref_3) have reported weak fields permeating amorphous solid water (ASW) condensed on substrates below 125$\,$K. Our results, presented in [Sup Mat Fig 2a](#sup_figure_2), reproduce the form of the temperature dependence of these Kelvin Probe measurements. The reports of electric fields permeating amorphous solid water attribute the residual polarization to unbalanced dipoles in pores,[4](#sup_ref_4),[5](sup_ref_5) and this effect can be recreated using molecular dynamics simulations.[6](#sup_ref_6)

A qualitative explanation for the variation of absorption wavelengths for data from ASW, between 20$\,$K and 125$\,$K in [Sup Mat Fig.2a](#sup_figure_2), rests upon theoretical estimates of the effect of an applied electric field on the energy of hydrogen bonds,[4](#sup_ref_4),[5](sup_ref_5) whose results may be summarized as follows: (i) local electric fields must be of the order of several x 10$^9$$\,$Vm$^{-1}$ to have an observable effect on hydrogen bond energy; (ii) such an electric field reduces the energy of hydrogen bonds when pointing in the same direction as the dipole moment of the system, as here, up to $\sim$100$\,$K; (iii) a switch in direction of the local electric field, implied by a switch in direction in the average field at >100$\,$K, results in a strengthening of the hydrogen bonds, consistent with the non-monotonic behaviour of absorption peak position around 100$\,$K seen in [Sup Mat Fig. 2a](#sup_figure_2).

**Figure 3 - Sup Mat** <a class="anchor" id="sup_figure_3"></a>: _The relative density of water ices deposited as a function of temperature. Material was prepared as thin films with thicknesses of 350$\,$nm. Ices were deposited in an ultra-high vacuum chamber with a base pressure below 2E-10$\,$mbar via condensation onto a KBr substrate, from a background partial pressure of 1E-6$\,$mbar (1E-5$\,$mbar at 150$\,$K) of pure water vapour. Data were collected at the same temperature as deposition except for films deposited at 150$\,$K, where measurement occurred at 140$\,$K to limit desorption. The relative density was estimated by dividing the band area of the O-H stretching mode in the infrared photoabsorption spectrum of each individual ice, by the thickness of the deposited ice. Ice thicknesses were measured using a laser interferometry technique.[7](#sup_ref_7) The IR absorbance spectrum for each ice (not shown) was measured in transmission geometry. The error bars in the figure, are smaller than individual data points and were estimated by quantifying the standard deviation of the O-H stretching mode band area for several depositions at 40$\,$K. Relative densities are shown assuming that the cross section for absorbance is the same across the entire temperature range._ 

These data show that ices increase in density as the deposition temperature rises until 120$\,$K. Above 120$\,$K, density remains constant. This represents the loss of porosity in amorphous ice films as deposition temperature is increased, until ices become crystalline with depositions above 120$\,$K. Density for crystalline films is constant with deposition temperature, indicating that crystalline films are not porous.
