```python
import numpy as np

DE = np.array([0.003381, 0.003381,0.003245,0.002990,0.002671,0.001978])
radius = 100
eps = 1.70712

result_1 = (1/96)
print('1 ',result_1)
result_2 = ((32*DE)/radius)
print('2 ',result_2)
result_3_num = ((16*(2**(1/3))*DE*(6+DE*eps*radius)))
print('3_num ',result_3_num)
                
result_3_den_i = radius**3 
print('3_den_i ',result_3_den_i)
result_3_den_ii = ((DE**2*eps**2 *radius**(0.5))*(-9+2*DE*eps*radius))
print('3_den_ii ',result_3_den_ii)
result_3_den_iii = 3*(3**(0.5))*(-DE**3 *eps**3 *(32+DE*eps*radius*(13+4*DE*eps*radius)))**0.5
print('3_den_iii ',result_3_den_iii)
result_3_den_iv = (radius**(9/2))
print('3_den_iv ',result_3_den_iv)

cube_root_part=((result_3_den_ii*result_3_den_iii)/result_3_den_iv)**(1/3)
print('cube_root_part ',cube_root_part)
second_argument = result_3_num/(result_3_den_i * (cube_root_part))
print('second argument ',second_argument)

result_4_num_i = 8*(2**(2/3))
print('4_num_i ',result_4_num_i)
result_4_num_ii = ((DE**2 *eps**2 *radius**(0.5))*(-9+2*DE*eps*radius))
print('4_num_ii ',result_4_num_ii)
result_4_num_iii = 3*(3**(0.5))*(-DE**3 *eps**3 *(32+DE*eps*radius*(13+4*DE*eps*radius)))**0.5
print('4_num_iii ',result_4_num_iii)
result_4_num_iv = (radius**(9/2))
print('4_num_iv ',result_4_num_iv)


third_argument = (result_4_num_i * cube_root_part**(1/3))/eps
print('third_argument ', third_argument)

result = result_1 *(result_2 + second_argument + third_argument)
print(result)
```
